## 1.Pie chart

```python

def piechart_bat(name):
    x=delivery[delivery['batsman']==name]
    y=x['batsman_runs'].value_counts()
    %matplotlib inline
    plt.pie(y,labels=y.index,autopct="%0.2f%%")
    plt.show()
    
```

```python
piechart_bat('Sachin Baby')
```

```python
x=delivery[delivery['batsman']=='V Kohli']
y=x['batsman_runs'].value_counts()
%matplotlib inline
plt.pie(y,labels=y.index,autopct="%0.2f%%")
plt.show()
```

#2.line chart

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=big[big['batsman']=='V Kohli']
batsman_data1=big[big['batsman']=='RG Sharma']
```

```python
perf=batsman_data.groupby('season').sum()['batsman_runs']
perf1=batsman_data1.groupby('season').sum()['batsman_runs']
```

```python
perf
```

```python
perf1
```

```python
plt.plot(perf.index,perf.values,color="green",marker='D',label="Virat Kohli")
plt.plot(perf1.index,perf1.values,color="red",marker='+',label="Rohit sharma")
plt.xlabel('season')
plt.ylabel('Runs scored')
plt.title('Batsman season by season performance')
plt.legend()
```

```python
def perfcomp(name,name1):
    batsman_data=big[big['batsman']==name]
    batsman_data1=big[big['batsman']==name1]
    perf=batsman_data.groupby('season').sum()['batsman_runs']
    perf1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(perf.index,perf.values,color="green",marker='D',label=name)
    plt.plot(perf1.index,perf1.values,color="red",marker='+',label=name1)
    plt.xlabel('season')
    plt.ylabel('Runs scored')
    plt.title('Batsman season by season performance')
    plt.legend()
```

```python
perfcomp('DA Warner','SR Watson')
```

# histogram

```python
run=delivery.groupby('batsman').sum()['batsman_runs']
```

```python
ball=delivery['batsman'].value_counts()
```

```python
sr=(run/ball)*100
```

```python
sr.sort_values(ascending=False)
```

```python
plt.hist(sr.values,bins=100)
plt.show()
```
